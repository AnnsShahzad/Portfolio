# format strings in plot function

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'g--')
```

Example format strings:

'b'    # blue markers with default shape

'or'   # red circles

'-g'   # green solid line

'--'   # dashed line with default color

'^k:'  # black triangle_up markers connected by a dotted line

Colors

The supported color abbreviations are the single letter codes

character	color

'b'	blue

'g'	green

'r'	red

'c'	cyan

'm'	magenta

'y'	yellow

'k'	black

'w'	white

```python
plt.plot(x,y,'k.-')
```

```python
plt.plot(x,y, color='black', marker='.', linestyle='-')
```

```python
plt.plot(x,y, color='black', marker='.', linestyle='-', markersize=1, linewidth=0.3)
```

```python
plt.plot(x,y, color='black', marker='.', linestyle='-',linewidth=10, alpha=0.1)  #alpha controls the transparency from 0-1
```

```python
plt.plot(x,y, color='black', marker='.', linestyle='-',animated=False)
```

#  Axes labels, Legend, Grid

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days =[1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.title('Max, Min, Avg Temperature over the days.')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.plot(days,max_t,color='black',marker='.',label='Max')
plt.plot(days,min_t,color='blue',marker='.',label='Min')
plt.plot(days,avg_t,color='red',marker='.',label='Avg')

plt.legend(loc='best',shadow=True)
```

```python
plt.title('Max, Min, Avg Temperature over the days.')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.plot(days,max_t,color='black',marker='.',label='Max')
plt.plot(days,min_t,color='blue',marker='.',label='Min')
plt.plot(days,avg_t,color='red',marker='.',label='Avg')

plt.legend(loc='best',shadow=True)
plt.grid()
```

# Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company=['Google','Amazone','Microsoft','Facebook','Kindle','Bing','Ebay','Wish']
revenue=[90,136,89,27,35,47,81,99]
profit=[30,36,9,10,17,13,33,47]
loss=[18,1,7,9,3,4,8,13]
```

```python
y_pos=np.arange(len(company))
y_pos
```

```python
plt.xticks(y_pos,company)
plt.ylabel('Revenue(billion)')
plt.title('US Tech Stocks')

plt.bar(y_pos,revenue, label='Revenue')
plt.bar(y_pos, profit, label='Profit')
plt.bar(y_pos, loss, label='Loss')

plt.legend(loc='best',shadow=True)
```

```python
plt.xticks(y_pos,company)
plt.ylabel('Revenue(billion)')
plt.title('US Tech Stocks')

plt.bar(y_pos-0.2,revenue,width=0.4, label='Revenue')
plt.bar(y_pos+0.2, profit,width=0.4, label='Profit')
plt.bar(y_pos+0.4, loss,width=0.4, label='Loss')

plt.legend(loc='best',shadow=True)
```

```python
plt.yticks(y_pos,company)
plt.xlabel('Revenue(billion)')
plt.title('US Tech Stocks')

plt.barh(y_pos,revenue, label='Revenue')
plt.barh(y_pos, profit, label='Profit')
plt.barh(y_pos, loss, label='Loss')

plt.legend(loc='best',shadow=True)
```
