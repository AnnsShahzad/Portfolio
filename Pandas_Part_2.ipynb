# Concatenate Dataframes

```python
import pandas as pd

pak_weather=pd.DataFrame({
    'City':['LHR','ISL','GRW','SIA'],
    'Temperature': [30,34,38,42],
    'Humidity':[80,78,76,74]
})
pak_weather
```

```python
US_weather=pd.DataFrame({
    'City':['NY','LA','LS','MIAMI'],
    'Temperature': [28,26,24,22],
    'Humidity':[66,62,60,58]
})
US_weather
```

```python
df=pd.concat([pak_weather,US_weather])
df
```

```python
df=pd.concat([pak_weather,US_weather], ignore_index=True)
df
```

```python
df=pd.concat([pak_weather,US_weather],keys=['PK','US'])
df
```

```python
windspeed=pd.DataFrame({
    'City':['NY','LA','LS','MIAMI'],
    'Wind Speed':[8,10,7,4]
})
windspeed
```

```python
temp=pd.DataFrame({
    'City':['NY','LA','LS','MIAMI'],
   'Temperature': [28,26,24,22]
})
windspeed
```

```python
df= pd.concat([temp,windspeed])
df
```

```python
df= pd.concat([temp,windspeed],axis=1)
df
```

```python
s = pd.Series(['Humid','Dry','Sunny','Cold'],name='Event')
s
```

```python
df = pd.concat([temp,windspeed,s],axis=1)
df
```
